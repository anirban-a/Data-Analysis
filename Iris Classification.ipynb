```python
from functools import partial
def amplified_min_max_scale(data):
    df = data.copy()
    def mod_min_max_scale(_max, _min, element):
        return (element-_min)/(_max - _min)
    
    for col in df.columns:
        _max, _min = df[col].max(), df[col].min()
        df[col] = df[col].apply(lambda x: mod_min_max_scale(_max, _min, x)**2)
    return df
```

```python
x1 = amplified_min_max_scale(X1)
```

```python
x2 = preprocessing.StandardScaler().fit_transform(X2)
```

```python
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1_train,x1_test,y1_train, y1_test = train_test_split(x1, y, test_size=0.25, random_state=43)
```

```python
x2_train,x2_test,y2_train, y2_test = train_test_split(x2, y, test_size=0.25, random_state=43)
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.model_selection import RandomizedSearchCV
```

```python
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt
```

```python
clf = DecisionTreeClassifier()
```

```python
clf1 = clf.fit(x1_train,y1_train)
```

```python
y1_pred = clf1.predict(x1_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy of Classifier 1:",metrics.accuracy_score(y1_test, y1_pred))
```

```python
clf2 = clf.fit(x2_train,y2_train)
```

```python
y2_pred = clf2.predict(x2_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy of Classifier 2:",metrics.accuracy_score(y2_test, y2_pred))
```

```python
param_dist = {"max_depth": [3, None],
              "max_features": sp_randInt(1, 5),
              "min_samples_leaf": sp_randInt(1, 5),
              "criterion": ["gini", "entropy"]}
```

```python
tree_cv = RandomizedSearchCV(clf, param_dist, cv=5)
```

```python
tree_cv.fit(x1_train, y1_train)
```

```python
print("Best score for Classifier 1 is {}".format(tree_cv.best_score_))
```

```python
tree_cv.fit(x2_train, y2_train)
print("Best score for Classifier 2 is {}".format(tree_cv.best_score_))
```

```python
tree_cv.fit(x1_train, y1_train)
print("Best score is {}".format(tree_cv.best_score_))
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_))
```

```python
tree = DecisionTreeClassifier(criterion=tree_cv.best_params_['criterion'], 
                              max_depth=tree_cv.best_params_['max_depth'], 
                              max_features=tree_cv.best_params_['max_features'],
                              min_samples_leaf=tree_cv.best_params_['min_samples_leaf'])
```

```python
tree.fit(x1_train, y1_train)
```

```python
y1_pred = tree.predict(x1_test)
# Model Accuracy, how often is the classifier correct?
print("Accuracy on X1:",metrics.accuracy_score(y1_test, y1_pred)*100)
```

```python
tree.fit(x2_train, y2_train)
```

```python
y2_pred = tree.predict(x2_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy on X2:",metrics.accuracy_score(y2_test, y2_pred)*100)
```
